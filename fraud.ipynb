```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties  # 匯入FontProperties

font = FontProperties(fname="/System/Library/Fonts/PingFang.ttc", size=15)
```

```python
fraud = pd.read_csv('/Users/chengchanglei/Library/Containers/com.microsoft.Excel/Data/Desktop/BANK_FRAUD_TWN_ALL1.csv')
fraud.index = fraud['年']
fraud = fraud[['詐欺筆數','詐欺金額[新台幣]']]
fraud
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
ax2 = plt.twinx()
l1, = ax1.plot(fraud['詐欺筆數'],color='#0080FF',marker='o')
l3, = ax2.plot(fraud['詐欺金額[新台幣]'],color = '#FF5809',marker='o')
ax1.grid()
plt.title('詐欺筆數 v.s 詐欺金額[新台幣]', fontsize = '20',fontproperties=font)
plt.legend([l1, l3], ['詐欺筆數', '詐欺金額[新台幣]'], loc = 2, prop=font)
plt.tight_layout()
```
